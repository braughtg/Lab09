```python
import numpy
from matplotlib.pyplot import *

year,co2 = numpy.loadtxt('emissions/historical_ghgs.csv', usecols=(0,1), delimiter=',', skiprows=1, unpack=True)

plot(year, co2, 'r-', label='C02')

legend(loc='upper left')                  # Add a legend

title("Historical GHG Emissions")  # Title for the plot.
xlabel("Year")                            # Label the x axis.
ylabel("Emissions pg")           # Label the y axis.
grid(True)                                # Turn on the grid lines.
show()                                    # Make the graph show up in the notebook.
```

```python
import numpy
from matplotlib.pyplot import *

x,y26 = numpy.loadtxt('emissions/rcp_2.6_data.csv', delimiter=',', skiprows=1, usecols=(0,7), unpack=True)
x,y45 = numpy.loadtxt('emissions/rcp_4.5_data.csv', delimiter=',', skiprows=1, usecols=(0,7), unpack=True)
x,y60 = numpy.loadtxt('emissions/rcp_6.0_data.csv', delimiter=',', skiprows=1, usecols=(0,7), unpack=True)
x,y85 = numpy.loadtxt('emissions/rcp_8.5_data.csv', delimiter=',', skiprows=1, usecols=(0,7), unpack=True)

plot(x, y26, 'r-', label='RCP 2.6')       # Plot a line.
plot(x, y45, 'b-', label='RCP 4.5')       # add another... 
plot(x, y60, 'g-', label='RCP 6.0')
plot(x, y85, 'y-', label='RCP 8.5')

legend(loc='upper left')                  # Add a legend

title("CO2 Concentration in RCP Models")  # Title for the plot.
xlabel("Year")                            # Label the x axis.
ylabel("CO2 Concentration ppm")           # Label the y axis.
grid(True)                                # Turn on the grid lines.
show()                                    # Make the graph show up in the notebook.
```

```python
import numpy
from matplotlib.pyplot import *

for trial in range(10):
  x,y26 = numpy.loadtxt('results/T' + '2.6' + '.'+str(trial)+'.csv', delimiter=',', skiprows=186, usecols=(2,30), unpack=True)
  x,y85 = numpy.loadtxt('results/T' + '8.5' + '.'+str(trial)+'.csv', delimiter=',', skiprows=186, usecols=(2,30), unpack=True)

  if trial == 0:
    plot(x, y26, 'g-', label='RCP 2.6')
    plot(x, y85, 'r-', label='RCP 8.5')
  else:
    plot(x, y26, 'g-')
    plot(x, y85, 'r-')
    
legend(loc='upper left')                  # Add a legend

title("Global Temperature Change")  # Title for the plot.
xlabel("Year")                            # Label the x axis.
ylabel("Temperature Change")           # Label the y axis.
grid(True)                                # Turn on the grid lines.
show()                                    # Make the graph show up in the notebook.
```
